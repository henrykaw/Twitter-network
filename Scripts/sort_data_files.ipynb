```python
def getFiles():
    return [file for file in os.listdir(path) if file.startswith("part-")]

def getMonthTimestamps(month):
    lastDayInMonth = calendar.monthrange(2016,month)[-1]
    firstTimestamp = time.mktime(datetime.datetime(2016,month,1,0,0,0).timetuple())
    lastTimestamp = time.mktime(datetime.datetime(2016,month,lastDayInMonth,23,59,59).timetuple())
    return firstTimestamp,lastTimestamp
list_of_hits = []

def hakuna(month):
    files = getFiles()
    list_of_hits = []
    file_count = 0
    fTimestamp, lTimestamp = getMonthTimestamps(month)
    for f in list(files):
        if file_count % 1000 == 0:
            print(f)
        file_count += 1
        with open(path+"/"+f) as file:
            for line in file:
                count = 0
                #elements = ast.literal_eval(line)
                elements = json.loads(line)
                for ts in elements[-1]:
                    if ts > fTimestamp and ts < lTimestamp:
                        count += 1
                if count > 0:
                    list_of_hits.append([elements[0][0],elements[0][1],count])
    matata(list_of_hits,month)

def matata(hits, month):
    with open(path_1+"/"+"csv_"+str(month)+".csv","w") as dump:
            writer = csv.writer(dump)
            #print(hits)
            writer.writerows(hits)
            print("Done")
        
```

```python
cProfile.run('hakuna(1)')
```
